```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# generate points for 2-d plane
a=np.array([1,2,3])
b=np.array([4,5,6,7])
a,b=np.meshgrid(a,b)
print(a)
print(b)
```

```python
# all point whose x belongs to a and y belongs to b form a 2-d plane
```

```python
fig=plt.figure(figsize=(10,10)) # not it has 0 axes
axes=fig.gca(projection="3d")
axes.plot_surface(a,b,a+b,cmap="coolwarm")# x=a,y=b,z=a+b
plt.show()
```

```python
aa=np.arange(-1,1,0.02)
bb=aa
a_x,b_x=np.meshgrid(aa,bb)
fig_1=plt.figure(figsize=(10,10))
axes=fig_1.gca(projection="3d")
axes.plot_surface(a_x,b_x,a_x**2+b_x**2,cmap="rainbow")
plt.show()
```

#### contour plot 

```python
fig_2=plt.figure(figsize=(10,10))
axes=fig_2.gca(projection="3d")
axes.contour(a_x,b_x,a_x**2+b_x**2,cmap="rainbow")
plt.show()
```
